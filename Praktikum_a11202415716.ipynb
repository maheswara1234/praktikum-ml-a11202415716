```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

class AnalisisDataNegara:
    def __init__(self, file_path):
        """Constructor untuk membaca file CSV saat objek dibuat"""
        print(f"Membaca data dari {file_path}...")
        # Membaca data menggunakan Pandas
        self.df = pd.read_csv(file_path)

        # Membersihkan data (Mengubah nama kolom 'Unnamed: 0' menjadi 'Kode')
        self.df.rename(columns={'Unnamed: 0': 'Kode'}, inplace=True)

    def tampilkan_tabel(self):
        """Method untuk menampilkan isi data tabular"""
        print("\n=== Tabel Data Negara ===")
        # Menampilkan tabel tanpa index default pandas agar lebih rapi
        display(self.df.style.hide(axis="index"))
        print("=========================\n")

    def visualisasi_populasi(self):
        """Method untuk menampilkan grafik batang Populasi Negara"""
        plt.figure(figsize=(10, 5))

        # Mengurutkan data berdasarkan populasi terbesar ke terkecil
        df_sorted = self.df.sort_values(by='Populasi', ascending=False)

        warna = ['#FF5722', '#2196F3', '#4CAF50', '#FFC107', '#9C27B0']
        bars = plt.bar(df_sorted['Negara'], df_sorted['Populasi'], color=warna)

        # Menambahkan teks nilai di atas setiap batang grafik
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{yval} Juta', ha='center', va='bottom', fontsize=10)

        plt.title('Perbandingan Populasi Berbagai Negara', fontsize=14, fontweight='bold', pad=15)
        plt.xlabel('Negara', fontsize=12)
        plt.ylabel('Populasi (Juta Jiwa)', fontsize=12)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.ylim(0, max(df_sorted['Populasi']) + 40) # Memberi ruang kosong di atas grafik
        plt.show()

    def visualisasi_area_direvisi(self):
        """Method untuk menampilkan grafik Pie Area Negara dengan label di luar (tanpa tumpang tindih)"""

        # 1. Menyiapkan Data
        labels = self.df['Negara']
        sizes = self.df['Area']
        colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0']

        # Menghitung persentase secara manual untuk label luar
        total = sum(sizes)
        percentages = [(i / total) * 100 for i in sizes]

        # 2. Membuat Figure dan Axes dengan ukuran lebih besar
        fig, ax = plt.subplots(figsize=(12, 10), subplot_kw=dict(aspect="equal"))

        # 3. Menambahkan 'Explode' (jarak) ke SEMUA irisan untuk kebersihan visual
        explode_data = (0.05, 0.05, 0.05, 0.05, 0.05)

        # 4. Membuat Pie Chart UTAMA (TIDAK menampilkan persentase di dalam)
        wedges, texts = ax.pie(sizes, explode=explode_data, labels=labels, colors=colors,
                                startangle=140, shadow=True,
                                wedgeprops={'linewidth': 1, 'edgecolor': 'white'},
                                textprops={'fontsize': 12}) # Ukuran font nama negara di luar

        # 5. Menambahkan Garis Penunjuk dan Label Persentase Manual di Luar
        bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
        kw = dict(xycoords='data', textcoords='data', arrowprops=dict(arrowstyle="-"),
                  bbox=bbox_props, zorder=0, va="center")

        for i, p in enumerate(wedges):
            ang = (p.theta2 - p.theta1)/2. + p.theta1
            y = np.sin(np.deg2rad(ang))
            x = np.cos(np.deg2rad(ang))

            # Pengontrolan posisi label luar
            horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
            connectionstyle = f"angle,angleA=0,angleB={ang}"
            kw["arrowprops"].update({"connectionstyle": connectionstyle})

            # Menambahkan anotasi persentase
            # Teks diletakkan lebih jauh dari lingkaran
            ax.annotate(f"{percentages[i]:.1f}%", xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                        horizontalalignment=horizontalalignment, **kw)

        ax.set_title('Persentase Luas Area Negara', fontsize=16, fontweight='bold', pad=25)
        plt.axis('equal') # Menjamin lingkaran sempurna
        plt.show()

# ==========================================
# BLOK EKSEKUSI UTAMA (Main Program)
# ==========================================

# 1. Instansiasi Objek dengan memasukkan nama file CSV
analisis = AnalisisDataNegara('Tab.csv')

# 2. Menampilkan Data Tabel
analisis.tampilkan_tabel()

# 3. Menampilkan Visualisasi Grafik Batang (Populasi)
analisis.visualisasi_populasi()

# 4. Menampilkan Visualisasi Grafik Pie (Direvisi, Tanpa Tumpang Tindih)
analisis.visualisasi_area_direvisi()
```
