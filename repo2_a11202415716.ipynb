```python
import pandas as pd
import numpy as np
import warnings

# Mengabaikan pesan peringatan (warning) saat perhitungan statistik
warnings.filterwarnings('ignore')

class AnalisisDataEPL:
    def __init__(self, file_path):
        """Constructor untuk memuat dan membersihkan dataset"""
        print(f"Mencoba memuat data dari: {file_path}")
        self.df = pd.read_csv(file_path)

        # Membersihkan data: Drop kolom 'Unnamed: 0' dan 'id' yang tidak bermakna numerik
        kolom_dibuang = ['Unnamed: 0', 'id']
        self.df_clean = self.df.drop(columns=[col for col in kolom_dibuang if col in self.df.columns])
        print("Data berhasil dimuat dan dibersihkan dari kolom ID yang tidak perlu.\n")

    def tampilkan_sampel_dan_statistik(self):
        """Menampilkan 5 data teratas dan statistik deskriptif"""
        print("=== 5 Baris Pertama Dataset ===")
        display(self.df_clean.head())

        print("\n=== Statistik Deskriptif (Kolom Numerik) ===")
        display(self.df_clean.describe())

    def deteksi_pencilan_assist(self):
        """Mencari pemain dengan jumlah assist di luar batas wajar (Outlier/Pencilan) menggunakan Tukey's Fences"""
        print("\n=== Pemain dengan Assist Terbanyak (Pencilan/Outlier) ===")

        # Ekstraksi hanya kolom numerik
        df_numeric = self.df_clean.select_dtypes(include=[np.number])

        # Logika Tukey's Fences
        q1 = df_numeric.quantile(0.25)
        q3 = df_numeric.quantile(0.75)
        iqr = q3 - q1

        # Menyelaraskan dataframe untuk operasi boolean
        df_align, iqr_new = df_numeric.align(iqr, axis=1, copy=False, join='outer')
        outlier_filter = (df_numeric < q1 - 1.5 * iqr_new) | (df_numeric > q3 + 1.5 * iqr_new)

        # Menerapkan filter khusus untuk kolom 'assists'
        pemain_assist_tinggi = self.df_clean[outlier_filter['assists']][['player_name', 'assists', 'team_title']]
        pemain_assist_tinggi = pemain_assist_tinggi.sort_values(by=['assists'], ascending=False)

        display(pemain_assist_tinggi)

    def analisis_performa_tim(self):
        """Menganalisis rata-rata gol per tim menggunakan groupby"""
        print("\n=== Rata-Rata Gol Pemain per Tim ===")
        rata_gol = self.df_clean.groupby('team_title')['goals'].mean().sort_values(ascending=False).reset_index()

        # Memperbaiki nama kolom agar lebih rapi
        rata_gol.columns = ['Nama Klub', 'Rata-Rata Gol']
        display(rata_gol)

    def uji_korelasi_pearson(self):
        """Menampilkan matriks korelasi Pearson antar variabel"""
        print("\n=== Matriks Korelasi Pearson ===")

        # Hanya menghitung korelasi pada kolom yang relevan (mengabaikan kolom non-numerik)
        korelasi = self.df_clean.corr(numeric_only=True)
        display(korelasi)

# ==========================================
# BLOK EKSEKUSI UTAMA (Main Program)
# ==========================================

# 1. Menentukan letak file CSV
path_file = "epl-goalScorer(20-21).csv"

# 2. Instansiasi Objek (Membuat alat analisisnya)
analisis_epl = AnalisisDataEPL(path_file)

# 3. Eksekusi metode-metode analisis secara berurutan
analisis_epl.tampilkan_sampel_dan_statistik()
analisis_epl.deteksi_pencilan_assist()
analisis_epl.analisis_performa_tim()
analisis_epl.uji_korelasi_pearson()
```
